```python
Q1 = df_raw['price_sq_m'].quantile(.001) 
Q2 = df_raw['price_sq_m'].quantile(.25) 
Q3 = df_raw['price_sq_m'].quantile(.75) 
Q4 = df_raw['price_sq_m'].quantile(.99)
Skewness1 = df_raw['price_sq_m'].skew()
Mean1 = df_raw['price_sq_m'].mean()
Median1 = df_raw['price_sq_m'].median()
IQR = Q3 - Q2
df = df_raw[(df_raw['price_sq_m'] > Q1)]
df = df_raw[(df_raw['price_sq_m'] < Q4)]
```

```python
streets = df.groupby('street')['price_sq_m'].agg(['mean'])
streets = streets.reset_index()
streets = streets.reset_index()
streets.rename(columns={'index':'street_index'}, inplace=True)  
df = pd.merge(df, streets[['street_index','street']], on='street', how='outer')
```

```python
districts = df.groupby('district')['price_sq_m'].agg(['mean'])
districts = districts.reset_index()
districts = districts.reset_index()
districts.rename(columns={'index':'district_index'}, inplace=True)  
df = pd.merge(df, districts[['district_index','district']], on='district', how='outer')
```

## Regression

### Single linear regression

* Let's test simple the regression of total price vs total space in the apartments.

```python
#code example:
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(df_ml[['space_sq_m']])
train_y = np.asanyarray(df_ml[['total_price']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
#plotting to show linear regression vs actual data:
plt.scatter(train.space_sq_m, train.total_price,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("space_sq_m")
plt.ylabel("total_price")
```

```python
#code example:
from sklearn.metrics import r2_score

test_x = np.asanyarray(df_ml[['space_sq_m']])
test_y = np.asanyarray(df_ml[['total_price']])
test_y_hat = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_hat - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_hat - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_hat , test_y) )
```

* the R2-score seems to be extremely low, making this model poor in efficiency. As expected, the space alone is a weak predictor of a price.

### Optimizing Decision Tree Regressor

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000,7000,10000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
# interestingly the model without specified max_leaf_nodes gets slightly better result than every selected
```

```python
best_tree_size = 10000

dtr_model_v2 = DecisionTreeRegressor(max_leaf_nodes=best_tree_size)

dtr_model_v2.fit(train_X, train_y)

val_predictions = dtr_model_v2.predict(val_X)

val_mae = mean_absolute_error(val_predictions, val_y)

print(val_mae)
```
